# Python String Methods: Split & Join

# 1. The `.split()` Method
The `.split()` method breaks a single string into a **list** of smaller strings based on a delimiter. By default, it splits by **spaces**.

# 2. The `.join()` Method 
It does the exact opposite of split. It takes a list of strings and glues them back together into a single string using a character of your choice.

```python
#If you receive data in that format and need to extract the number, 
# You can use the .split() method to separate the text by spaces and grab just the 
#first part ("100") before converting it:
value = "100 followers"

# .split() turns "100 followers" into ['100', 'followers']
# [0] grabs the first item '100'
value = int(value.split()[0]) 

print(value)  # Output: 100
```

```python
value = "100 followers"

# Grab characters from index 0 up to 3 (which is "100")
just_the_number = value[:3] 

# Cast it safely
print(int(just_the_number))  # Output: 100
```

```python
#Split
text = "100 followers"
result = text.split()

print(result) 
# Output: ['100', 'followers']
```

```python
# Splitting with different delimiters
# Default delimiter is space

csv_data = "apple,banana,cherry,date"
# Split at every comma
fruits = csv_data.split(",")
print(fruits)


log_line = "ERROR#404#Page Not Found#10:15AM"
# Split at every hashtag
log_details = log_line.split("#")
print(log_details)
print(f"Error Code: {log_details[1]}") 


date_string = "03/06/2026"
# Split at every forward slash
date_pieces = date_string.split("/")
print(date_pieces)
# Output: ['03', '06', '2026']
```

```python
#Changing the Delimiter Completely (Split + Join)
#If your goal is to take a string that uses one delimiter (like spaces)
#and change it to use a different one (like hyphens), you combine .split() with .join().

sentence = "Python is fun to learn"

# Step 1: Split by spaces to get a list
words = sentence.split() 

# Step 2: Glue them back together using a hyphen
new_sentence = "-".join(words)

print(new_sentence)
# Output: Python-is-fun-to-learn
```
