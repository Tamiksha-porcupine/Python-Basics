## Advanced Indexing: Finding the Second Occurrence

When working with structured strings like `"ID:9482-NAME:Alex"`, a standard `.index()` call will only return the first occurrence of a character. If we need to find a subsequent occurrence (like the second colon), we can approach it in two distinct ways.

### Approach 1: Using `.rindex()` (Right-to-Left Search)
The `.rindex()` method works exactly like `.index()`, but it starts scanning the string from the **right side** (the end) instead of the left side. 

* **Best used when:** You specifically need the *last* occurrence of a character in a string.
* **Pros:** Clean, short, and requires only a single line of code.
* **Cons:** If there are three or more colons, this will always jump straight to the final one, missing the middle ones.


```python
data = "ID:9482-NAME:Alex"
print(data[data.rindex(':') + 1])
```

```python
data = "ID:9482-NAME:Alex"

# Find the first colon starting from the right side
second_colon = data.rindex(':') 

# Grab the character right after it
print(data[second_colon + 1])  # This will output 'A'
```

```python
data = "ID:9482-NAME:Alex"

# 1. Find the first colon's position (which is 2)
first_colon = data.index(':')

# 2. Find the second colon by starting the search at first_colon + 1
second_colon = data.index(':', first_colon + 1)
print(second_colon)  # This will output 12

# 3. Get the character right after it
print(data[second_colon + 1])  # This will output 'A'
```
